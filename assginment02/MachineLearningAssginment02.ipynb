```
from matplotlib import pyplot as plt
import numpy as np
import sympy as sym
import random

#Initialize input data
m = 100
x = np.random.random(m) * 10
yHat = (2*x) + 2
y = yHat + np.random.randn(len(y))
print(y)
```

```
#initialize variables
theta0 = 10
theta1 = 1
```

```
#Define linear function & cost function & derivative function of cost function
def hTheta():
  return theta1*x + theta0

def j():
  return (np.sum((hTheta() - y)**2)/(2*m))

def jPrimeTheta0():
  return (np.sum(hTheta() - y))/m

def jPrimeTheta1():
  return (np.sum((hTheta() -y)*x))/m

#Visualize
plt.scatter(x, y, color=[0, 0, 0])
plt.title('Before Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, yHat)
plt.show()
```

```
#Define update function

learningRate = 0.001

def updateTheta():
  updateTheta0 = theta0 - (learningRate*jPrimeTheta0())
  updateTheta1 = theta1 - (learningRate*jPrimeTheta1())

  return updateTheta0, updateTheta1
```

```
#Energy value, Theta0, Theta1 for visualization
energyValue=[]
theta0Value=[]
theta1Value=[]

#Learning
while(1):
  energyValue.append(j())
  theta0Value.append(theta0)
  theta1Value.append(theta1)

  tempTheta0 = theta0
  tempTheta1 = theta1
  theta0, theta1 = updateTheta()
  if (tempTheta0 == theta0) and (tempTheta1 == theta1):
    break

#Visualize
plt.scatter(x, y, color = [0, 0, 0])
plt.title('After Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, hTheta(), color=[1, 0, 0])
plt.show()
print(theta0, theta1)
```

```
#Visualize Energy Value
plt.scatter(range(len(energyValue)), energyValue, color = [0, 0, 1])
plt.title('Energy Value')
plt.xlabel('#Iteration')
plt.ylabel('Energy Value')
plt.show()
```

```
#Visualize Energy Value
plt.scatter(range(len(theta0Value)), theta0Value, color = [1, 0, 0])
plt.scatter(range(len(theta1Value)), theta1Value, color = [0, 0, 1])
plt.title('Theta0 & Theta1 Values')
plt.xlabel('#Iteration')
plt.ylabel('Theta0 & Theta1')
plt.show()
```
