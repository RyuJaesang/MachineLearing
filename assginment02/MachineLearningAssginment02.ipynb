```
from matplotlib import pyplot as plt
import numpy as np
import sympy as sym

#Initialize input data
m = 100
x = np.random.random(m) * 10
yHat = (2*x) + 2
y = yHat + np.random.random(yHat.shape)
```

```
#initialize variables
theta0 = 10
theta1 = 1
```

```
#Define linear function & cost function
def hTheta():
  return theta1*x + theta0

def j():
  return (np.sum((hTheta() - y)**2)/(2*m))

def jPrimeTheta0():
  return (np.sum(hTheta() - y))/m

def jPrimeTheta1():
  return (np.sum((hTheta() -y)*x))/m

#Visualize
plt.scatter(x, y, color=[0, 0, 0])
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, hTheta())
plt.show()
```

```
#Define update function

learningRate = 0.001

def updateTheta():
  updateTheta0 = theta0 - (learningRate*jPrimeTheta0())
  updateTheta1 = theta1 - (learningRate*jPrimeTheta1())

  return updateTheta0, updateTheta1
```

```
#Energy value
energyValue=[]

#Learning
while(1):
  tempTheta0 = theta0
  tempTheta1 = theta1
  energyValue.append(j())
  theta0, theta1 = updateTheta()
  if (tempTheta0 == theta0) and (tempTheta1 == theta1):
    break

#Visualize
plt.scatter(x, y, color = [0, 0, 0])
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, hTheta(), color=[1, 0, 0])
plt.show()
print(theta0, theta1)
```

```
#Visualize Energy Value
plt.scatter(range(len(energyValue)), energyValue, color = [0, 0, 1])
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
print(theta0, theta1)
```
