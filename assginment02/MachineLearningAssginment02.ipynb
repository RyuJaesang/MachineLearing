```
from matplotlib import pyplot as plt
import numpy as np
import sympy as sym

#Initialize input data
m = 100
x = np.random.random(m) * 10
y = (2*x) + 2
yHat = y + np.random.random(y.shape)

#Visualize
plt.plot(x, yHat, 'ro')
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```
#initialize variables
theta0 = sym.symbols('theta0')
theta1 = sym.symbols('theta1')

theta0 = 10
theta1 = 1
```

```
#Define linear function
def hTheta(x): 
  return theta1*x + theta0

def j(theta0, theta1):
  return (np.sum(((theta1*x + theta0) - y)**2)/(2*m))

def jPrimeTheta0(theta0, theta1):
  return (theta1*x + theta0 - y)

  
#Visualize
plt.plot(x, yHat, 'ro')
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, hTheta(x))
plt.show()
```
