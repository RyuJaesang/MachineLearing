```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

data = np.genfromtxt(file, delimiter=',')

x = data[:, 0]
y = data[:, 1]
m = len(x)

#Visualization
plt.figure(figsize=(8, 8))
plt.scatter(x, y, alpha=0.3, color = [0, 0, 0])
plt.show()
```

```
#initialize variables
theta0 = -30
theta1 = -30
```

```
#Define linear function & cost function & derivative function of cost function
def hTheta():
  return theta1*x + theta0

def j():
  return (np.sum((hTheta() - y)**2)/(2*m))

def jPrimeTheta0():
  return (np.sum(hTheta() - y))/m

def jPrimeTheta1():
  return (np.sum((hTheta() -y)*x))/m
```

```
#Define update function

learningRate = 0.001

def updateTheta():
  updateTheta0 = theta0 - (learningRate*jPrimeTheta0())
  updateTheta1 = theta1 - (learningRate*jPrimeTheta1())

  return updateTheta0, updateTheta1
```

```
#Energy value, Theta0, Theta1 for visualization
energyValue=[]
theta0Value=[]
theta1Value=[]

#Learning
while(1):
  energyValue.append(j())
  theta0Value.append(theta0)
  theta1Value.append(theta1)

  tempTheta0 = theta0
  tempTheta1 = theta1
  theta0, theta1 = updateTheta()
  if (tempTheta0 == theta0) and (tempTheta1 == theta1):
    break
```

```
#Visualize
plt.figure(figsize=(8, 8))
plt.scatter(x, y, alpha=0.3, color = [0, 0, 0])
plt.title('After Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, hTheta(), color=[1, 0, 0])
plt.show()
print(theta0, theta1)
```

```
def jVisualize(local_theta0, local_theta1):
  for i in range(len(x)):
    return np.sum((local_theta1*x + local_theta0 - y)**2/(2*m))
tempJ = []
arangeTheta0 = np.arange(-30, 30, 0.1)
arangeTheta1 = np.arange(-30, 30, 0.1)
for i in range(len(arangeTheta1)):
  for j in range(len(arangeTheta0)):
   tempJ.append(jVisualize(arangeTheta0[j], arangeTheta1[i]))
arangeTheta0, arangeTheta1 = np.meshgrid(arangeTheta0, arangeTheta1)
# z = np.sum((arangeTheta1*x + arangeTheta0 - y)**2/(2*m))
arangeJ = np.array(tempJ).reshape(600,600)
```

```
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(arangeTheta0, arangeTheta1, arangeJ, shade=True, alpha = 0.5)
ax.set_xlabel('Theta0')
ax.set_ylabel('Theta1')
ax.set_zlabel('Energy Value')
plt.title('Surface')
plt.show()
```
