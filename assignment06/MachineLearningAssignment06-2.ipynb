```
import numpy as np
import math
import matplotlib.pyplot as plt
import sympy

with open('drive/My Drive/machine_learning_data/assignment06/data-nonlinear.txt', newline='') as file:
  data= np.genfromtxt(file, delimiter=',')

x = data[:, 0]
y = data[:, 1]
label = data[:, 2]
m = len(x)

x0 = x[label == 0]
y0 = y[label == 0]

x1 = x[label == 1]
y1 = y[label == 1]

plt.figure()
plt.scatter(x0, y0, c='b')
plt.scatter(x1, y1, c='r')
plt.tight_layout()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```
#initialize Train Parameter
theta0 = 1
theta1 = 1.1
theta2 = 1.2
theta3 = 1.3
theta4 = 1.4
theta5 = 1.5
theta6 = 1.6
theta7 = 1.7
theta8 = 1.8
theta9 = 1.9
theta10 = 2
theta11 = 2.1
theta12 = 2.2
theta13 = 2.3
theta14 = 2.4
```

```
#Define linear function & cost function & derivative function of cost function
def zFunc(x, y):
  return theta0 + theta1*x + theta2*y + theta3*x*y + theta4*x*x + theta5*y*y + theta6*x*x*x + theta7*x*x*y + theta8*x*y*y + theta9*y*y*y + theta10*x*x*x*x + theta11*x*x*x*y + theta12*x*x*y*y + theta13*x*y*y*y + theta14*y*y*y*y

def sigmaFunc(x, y):
  return 1/(1+np.exp(-1 * zFunc(x, y)))

def j(x, y, label):
  return 1/m * np.sum(((-1 * label) * np.log(sigmaFunc(x, y))) - ((1-label) * np.log(1-sigmaFunc(x, y))))

def jPrimeTheta0(x, y, label):
  return 1/m * np.sum(sigmaFunc(x, y) - label)

def jPrimeTheta1(x, y, label):
  return 1/m * np.sum((sigmaFunc(x, y) - label) * x)

def jPrimeTheta2(x, y, label):
  return 1/m * np.sum((sigmaFunc(x, y) - label) * y)

def jPrimeTheta3(x, y, label):
  return 1/m * np.sum((sigmaFunc(x, y) - label) * x * y)

def jPrimeTheta4(x, y, label):
  return 1/m * np.sum((sigmaFunc(x, y) - label) * x * x)

def jPrimeTheta5(x, y, label):
  return 1/m * np.sum((sigmaFunc(x, y) - label) * y * y)

def jPrimeTheta6(x, y, label):
  return 1/m * np.sum((sigmaFunc(x, y) - label) * x * x * x)

def jPrimeTheta7(x, y, label):
  return 1/m * np.sum((sigmaFunc(x, y) - label) * x * x * y)

def jPrimeTheta8(x, y, label):
  return 1/m * np.sum((sigmaFunc(x, y) - label) * x * y * y)

def jPrimeTheta9(x, y, label):
  return 1/m * np.sum((sigmaFunc(x, y) - label) * y * y * y)

def jPrimeTheta10(x, y, label):
  return 1/m * np.sum((sigmaFunc(x, y) - label) * x * x * x * x)

def jPrimeTheta11(x, y, label):
  return 1/m * np.sum((sigmaFunc(x, y) - label) * x * x * x * y)

def jPrimeTheta12(x, y, label):
  return 1/m * np.sum((sigmaFunc(x, y) - label) * x * x * y * y)

def jPrimeTheta13(x, y, label):
  return 1/m * np.sum((sigmaFunc(x, y) - label) * x * y * y * y)

def jPrimeTheta14(x, y, label):
  return 1/m * np.sum((sigmaFunc(x, y) - label) * y * y * y * y)
```

```
# Define update Function

learningRate = 0.005

def updateTheta(x, y, label):
  updateTheta = []
  updateTheta.append(theta0 - (learningRate * jPrimeTheta0(x, y, label)))
  updateTheta.append(theta1 - (learningRate * jPrimeTheta1(x, y, label)))
  updateTheta.append(theta2 - (learningRate * jPrimeTheta2(x, y, label)))
  updateTheta.append(theta3 - (learningRate * jPrimeTheta3(x, y, label)))
  updateTheta.append(theta4 - (learningRate * jPrimeTheta4(x, y, label)))
  updateTheta.append(theta5 - (learningRate * jPrimeTheta5(x, y, label)))
  updateTheta.append(theta6 - (learningRate * jPrimeTheta6(x, y, label)))
  updateTheta.append(theta7 - (learningRate * jPrimeTheta7(x, y, label)))
  updateTheta.append(theta8 - (learningRate * jPrimeTheta8(x, y, label)))
  updateTheta.append(theta9 - (learningRate * jPrimeTheta9(x, y, label)))
  updateTheta.append(theta10 - (learningRate * jPrimeTheta10(x, y, label)))
  updateTheta.append(theta11 - (learningRate * jPrimeTheta11(x, y, label)))
  updateTheta.append(theta12 - (learningRate * jPrimeTheta12(x, y, label))) 
  updateTheta.append(theta13 - (learningRate * jPrimeTheta13(x, y, label)))
  updateTheta.append(theta14 - (learningRate * jPrimeTheta14(x, y, label)))

  return updateTheta
```

```
#Learning
trainingError = []
accuracy = []
counter = 0
heu_accuracy = 101

while(1):
  counter = counter+1
  trainingError.append(j(x, y, label))
  tempTrainingError = j(x, y, label)

  tempTheta = []
  tempTheta.append(theta0)
  tempTheta.append(theta1)
  tempTheta.append(theta2)
  tempTheta.append(theta3)
  tempTheta.append(theta4)
  tempTheta.append(theta5)
  tempTheta.append(theta6)
  tempTheta.append(theta7)
  tempTheta.append(theta8)
  tempTheta.append(theta9)
  tempTheta.append(theta10)
  tempTheta.append(theta11)
  tempTheta.append(theta12)
  tempTheta.append(theta13)
  tempTheta.append(theta14)

  thetaUpdate = updateTheta(x, y, label)
  theta0 = thetaUpdate[0]
  theta1 = thetaUpdate[1]    
  theta2 = thetaUpdate[2]    
  theta3 = thetaUpdate[3]    
  theta4 = thetaUpdate[4]    
  theta5 = thetaUpdate[5]    
  theta6 = thetaUpdate[6]
  theta7 = thetaUpdate[7]    
  theta8 = thetaUpdate[8]    
  theta9 = thetaUpdate[9]    
  theta10 = thetaUpdate[10]    
  theta11 = thetaUpdate[11] 
  theta12 = thetaUpdate[12]
  theta13 = thetaUpdate[13]    
  theta14 = thetaUpdate[14]    

  ### Print part
  if (counter%1000) == 0:
    temp_accuracy = 0
    for i in range(len(x)):
      if ((sigmaFunc(x[i], y[i]) > 0.5) and (label[i] == 1)) or ((sigmaFunc(x[i], y[i]) < 0.5) and (label[i] == 0)):
        temp_accuracy = temp_accuracy + 1
    accuracy.append(temp_accuracy)
    print('Accuracy : ' + str(temp_accuracy))
    print('Training Error : ' + str(j(x, y, label)))

    if temp_accuracy == heu_accuracy:
      break
```
